```python
import h5py
import numpy as np
from itertools import combinations
from sklearn.cluster import KMeans
```

```python
def load_h5_file(file, num_cost_matrices):
    """Loads the train and test embeddings"""
    with h5py.File(file, 'r') as f:
        cost_matrices = {i:np.array(f[f'cost_matrix_{i}']) for i in range(num_cost_matrices)}
    return cost_matrices
```

```python
len_dataset = 188
num_cost_matrices = len(list(combinations(range(len_dataset), r=2)))

print(f'Number of cost matrices: {num_cost_matrices}')

cost_matrices = load_h5_file('../data/cost_matrices.h5', num_cost_matrices)
```

```python
cost_matrices[0]
```

```python
row_n_clusters = 2
col_n_clusters = 2
kmeans_rows = KMeans(n_clusters=row_n_clusters)
kmeans_cols = KMeans(n_clusters=col_n_clusters)
```

```python
kmeans_rows.fit(cost_matrices[0])
kmeans_cols.fit(cost_matrices[0].T)
row_clusters = kmeans_rows.labels_
col_clusters = kmeans_cols.labels_
```

```python
sub_matrices = []
for i in range(num_cost_matrices):
    cluster_matrices = []
    kmeans_rows = KMeans(n_clusters=row_n_clusters)
    kmeans_rows.fit(cost_matrices[i])
    row_clusters = kmeans_rows.labels_
    kmeans_cols = KMeans(n_clusters=col_n_clusters)
    kmeans_cols.fit(cost_matrices[i].T)
    col_clusters = kmeans_cols.labels_
    row_ids = np.arange(row_n_clusters)
    col_ids = np.arange(col_n_clusters)
    for row_id in row_ids:
        for col_id in col_ids:
            row_indices = np.where(row_clusters == row_id)[0]
            col_indices = np.where(col_clusters == col_id)[0]

            sub_matrix = cost_matrices[0][row_indices, :][:, col_indices]
            cluster_matrices.append(sub_matrix)
    sub_matrices.append(cluster_matrices)
```
